```python
import pandas as pd

def extract_data_from_google_sheet(SHEET_ID: str):
    URL = f"https://docs.google.com/spreadsheets/d/{SHEET_ID}/export?format=csv"

    df = pd.read_csv(URL)
    return df
```

```python
# Ensure the column is integer type
df['SC #'] = df['SC #'].astype('Int64')  # nullable integer type in pandas
```

```python
df.rename(columns={
    'SC #': 'Segment ID',
}, inplace=True)
```

```python
df.info()
```

```python
df[['Segment Name']].head(10)
```

```python
df['Segment Name'] = df['Segment Name'].str.strip()
df['Segment Name'] = df['Segment Name'].str.replace(r'\s+', ' ', regex=True)
import re

df['Segment Name'] = df['Segment Name'].apply(lambda x: re.sub(r'[^\w\s\+\-&]', '', x))
```

```python
print(df['Segment Name'].isna().sum())
```

```python
df.columns = df.columns.str.strip()  # Remove leading/trailing whitespace from column names
```

```python
# Fill missing values with a default, e.g., "Uncategorized"
df['Segment Folder'] = df['Segment Folder'].fillna('Uncategorized')
```

```python
df['Segment Folder'] = df['Segment Folder'].str.strip()
```
