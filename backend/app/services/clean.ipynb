```python
import ast

def safe_list_parse(x):
    # If it's a string like "[46, 141]", convert to list
    if isinstance(x, str):
        try:
            return [int(i) for i in ast.literal_eval(x)]
        except Exception:
            return []  # fallback for malformed strings
    # If it's already a list, make sure all elements are integers
    elif isinstance(x, list):
        return [int(i) for i in x if i is not None]
    else:
        return []  # fallback for NaN or other types

df['Audience Segment A IDs'] = df['Audience Segment A IDs'].apply(safe_list_parse)
```

```python
import json
import ast

def clean_audience_column(x):
    # If it's a string like "[46, 141]", convert to list
    if isinstance(x, str):
        try:
            return json.dumps([int(i) for i in ast.literal_eval(x)])
        except Exception:
            return json.dumps([])  # fallback for malformed strings
    # If it's already a list, convert elements to int and dump as JSON string
    elif isinstance(x, list):
        return json.dumps([int(i) for i in x if i is not None])
    else:
        # fallback for NaN or other types
        return json.dumps([])

# Apply to both columns
df_cleaned_2['Audience Segment A IDs'] = df_cleaned_2['Audience Segment A IDs'].apply(clean_audience_column)
df_cleaned_2['Audience Segment B IDs'] = df_cleaned_2['Audience Segment B IDs'].apply(clean_audience_column)

# Check first rows
print(df_cleaned_2[['Audience Segment A IDs', 'Audience Segment B IDs']].head(100))
```
